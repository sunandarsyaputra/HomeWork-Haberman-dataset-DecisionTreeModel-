# Import  Library

```python
import pandas as pd
import numpy as np 
import pandas as pd 
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
```

# Collect data

```python
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.csv'
list_cols = ['Age', 'Patients Years', 'N_positive_ax', 'Survival_status']
df = pd.read_csv(url, names= list_cols)

df.head()
```

# 1. Bagi kedua data ini menjadi data training dan data test dengan test_size=0.25.

```python
X = df.drop(['Survival_status'],1)
y = df['Survival_status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =  train_test_split(X, y, test_size = 0.25,random_state = 123)
```

# 2. Pelajari tentang metrics roc_auc_score kemudian buatlah model dan evaluasi dengan menggunakan teknik cross-validation dengan scoring 'roc_auc'.

```python
from sklearn.tree import DecisionTreeClassifier
dectree = DecisionTreeClassifier(criterion = 'entropy', random_state = 123, max_depth=3)
dectree.fit(X_train, y_train)
```

```python
## Accuracy Score
from sklearn.metrics import accuracy_score
y_test_pred = dectree.predict(X_test)
accuracy_score(y_test, y_test_pred)
```

```python
## Cross validasi Score
from sklearn.model_selection import cross_val_score
print(cross_val_score(dectree, X, y, cv=3))
```

```python
## Roc-Auc Score
from sklearn.metrics import roc_auc_score
roc_auc_score(y, dectree.predict_proba(X)[:, 1])
```

# 3. Berapa score rata-rata dari model dengan teknik cross-validation tersebut? 

```python
## Cross validasi Score
from sklearn.model_selection import cross_val_score
print(cross_val_score(dectree, X, y, cv=3).mean())
```

# 4. Prediksi data test dengan model yang telah kalian buat! (coding)

```python
prediction = dectree.predict(X_test)
```

```python
predik_survived = pd.DataFrame(data= prediction, columns= ['Prediksi Survived']).reset_index()
predik_survived
```

# 5. Bagaimana hasil confusion matrix dari hasil prediksi tersebut? (coding)

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
predictions = cross_val_predict(dectree, X_test, y_test, cv=3)
confusion_matrix(y_test, predictions)
```

# 6. Bagaimana classification report dari hasil prediksi tersebut?

```python
from sklearn.metrics import precision_score, recall_score

print("Precision:", precision_score(y_test, predictions))
print("Recall:",recall_score(y_test, predictions))
```
